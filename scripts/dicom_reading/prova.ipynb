```python
import os
import pydicom
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Specify the path to your DICOM folder
#dicom_folder_path = 'D:\\script_di_ale\\dicoms\\2122818'  # Notice the double backslashes for Windows paths
dicom_folder_path = r"D:\script_di_ale\dicoms"
excel_path = r"D:\script_di_ale\I3LUNG_CT_series_number.xlsx"
df = pd.read_excel(excel_path)

#show lines where CT_series_number is 5,6 or 7 and SEG_series_number is '1001', '1004', '1975', '1052', '1054'
df2 = df[(df['CT_series_number'].isin([9,14]))]
df2 = df2[(df2['SEG_series_number'].isin([1001, 1027]))]
dicom_folder_path += '\\' + df2['I3LUNG_ID'].iloc[0]
print(dicom_folder_path)
# List to hold all the DICOM datasets
dicom_datasets = {}
print(os.listdir(dicom_folder_path))
# Loop through all the files in the folder
try:
        os.remove(os.path.join(dicom_folder_path, 'desktop.ini'))
        print("tolto lo stronzo")
        
except:
        print('ah... t\'appost')
    
for filename in os.listdir(dicom_folder_path):
    file_path = os.path.join(dicom_folder_path, filename)
    # Read the DICOM file
    ds = pydicom.dcmread(file_path)
    # Append the dataset to the list
    dicom_datasets[filename] = ds
```

```python
import os
import tempfile
import shutil
import dicom2nifti
import nibabel as nib
#dicom_files is files with series 9


# Create a temporary directory
temp_dir_dicom = tempfile.mkdtemp()
temp_dir_out = tempfile.mkdtemp()
nifti_output_path = r"D:\script_di_ale"

# Copy the DICOM files to the temporary directory
for file_path in dicom_files:
    file_path = os.path.join(dicom_folder_path, file_path)
    shutil.copy(file_path, temp_dir_dicom)

# Specify the NIfTI file output path in the current working directory


# Convert the DICOM files in the temporary directory to a NIfTI file
dicom2nifti.convert_directory(temp_dir_dicom, temp_dir_out, compression=True, reorient=True)
#take the file in temp out and rename it
for filename in os.listdir(temp_dir_out):
    file_path = os.path.join(temp_dir_out, filename)
    shutil.copy(file_path, nifti_output_path)
    os.rename(nifti_output_path + '\\' + filename, nifti_output_path + '\\' + "fighissimo.nii.gz")



# Load the NIfTI file and inspect the affine matrix and other details
nifti_img = nib.load(nifti_output_path + '\\' + "fighissimo.nii.gz")
print("Affine matrix:")
print(nifti_img.affine)

# Display additional information as needed, e.g., image shape
print("Image shape:", nifti_img.shape)

# Clean up: Remove the temporary directory and its contents when done
shutil.rmtree(temp_dir_dicom)
shutil.rmtree(temp_dir_out)
```

```python
import subprocess
import logging

os.sep = "/"
volume, pseudonym = "D:" + os.sep + "script_di_ale", os.sep + "tmp" + os.sep + "script_di_ale"
output_labels = pseudonym + os.sep + "output"
def run_docker_command(input_path, name, format="nifti"):
    # Corrected Docker command with volume mapping and paths adjusted for Windows
    command = [
        "docker", "run", 
        "-v", f"{volume}:{pseudonym}", 
        "qiicr/dcmqi", 
        "segimage2itkimage", 
        "-t", format,
        "-p", name,
        "--outputDirectory", output_labels, 
        "--inputDICOM", input_path
    ]

    # Run the command
    print("should be running")
    print(f"Running Docker command: {' '.join(command)}")
    result = subprocess.run(command, stdout=subprocess.PIPE, stderr=subprocess.PIPE, text=True)
    if result.returncode == 0:

        logging.info("Docker command executed successfully!")
        logging.info(f"Output:\n{result.stdout}")
    else:
        logging.error("Error running Docker command:")
        logging.error(f"Error output:\n{result.stderr}")

```

```python
def find_first_difference(str1, str2):
    # Iterate over the shorter of the two strings
    for index in range(min(len(str1), len(str2))):
        # Check if characters at the current position are different
        if str1[index] != str2[index]:
            # Return the index of the first difference
            return index
    # If no difference has been found within the length of the shorter string,
    # and the strings are of different lengths, return the length of the shorter string.
    # Otherwise, return None as the strings are identical.
    if len(str1) != len(str2):
        return min(len(str1), len(str2))
    return None  # The strings are identical

# Example usage
index = find_first_difference("hello", "helicopter")
print(index)  # This will print the index of the first differing character
```
